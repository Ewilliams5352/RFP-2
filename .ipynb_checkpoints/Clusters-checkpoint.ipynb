```python
def get_song_genre(song_name):
    # Search for the song and get the artist's ID
    results = sp.search(q='track:' + song_name, type='track')
    items = results['tracks']['items']
    if not items:
        return "Song not found."
    
    artist_id = items[0]['artists'][0]['id']
    
    # Use the artist's ID to fetch the artist's genres
    artist = sp.artist(artist_id)
    genres = artist['genres']
    
    if genres:
        return genres
    else:
        return "Genre not found."

song_genres = []
counter = 0
for song_name in songs_name_list:
    song_genres.append(get_song_genre(song_name))
    if counter == 1499:
        break
    counter+=1
```

```python
clustering_songs_df["genre"] = song_genres
```
